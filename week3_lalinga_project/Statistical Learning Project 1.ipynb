```python
#Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (7 points)
def Goal_diff_count():
    return laligaCup["GoalsFor"] - laligaCup["GoalsAgainst"]
```
